## 凡例 ・ タイトル
---

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 25, 100)  # 0~25(等間隔)、0行100列のデータ作成

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, "-b", label="sine")
plt.plot(x, y2, "-r", label="cosine")

# i) 凡例 （legend）　※ちっちゃくグラフ説明してるやつ
plt.legend(loc='upper left')  # loc(location)
plt.ylim(-1.5, 2.0)  # ylim()関数：Y軸幅調整
# plt.show()


# ii) タイトル　グリッド線
# ※一度show()をしてしまったので再度値をインサートしてから
plt.title('Sine & Cosine Plot')
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.grid(True)
plt.show()


# iii) 関数化
def sq(i):
    return i**2

def cube(i):
    return i**3

def power4(i):
    return i**4

plt.plot(x , sq(x), 'r')
plt.plot(x , cube(x), 'b')
plt.plot(x , power4(x), 'y')
plt.show
````

## サブプロット
---
+ サブプロット(メソッド)

      2つのデータセットを比較したい時
      グラフを複数個いっぺんに描画させたい時

````python
import matplotlib.pyplot as plt
import numpy as np

v = np.linspace(-np.pi, np.pi, 100)  # -314~3.14(Π) 等間隔で１００個

x1 = np.sin(2*v)
x2 = np.cos(2*v)

# サブプロットメソッド
figure,(Left, Right) = plt.subplots(ncols=2, figsize=(10, 10))  # 2つの行指定 figsize(縦、横)

Left.plot(v, x1, linewidth=2)
Left.set_title('sine')
Left.set_xlabel('Value')
Left.set_ylabel('X')
Left.set_xlim(-np.pi, np.pi)
Left.grid(True)

Right.plot(v, x2, linewidth=2)
Right.set_title('cosin')
Right.set_xlabel('Value')
Right.set_ylabel('X')
Right.set_xlim(-np.pi, np.pi)
Right.grid(True)

figure.show()
````

## 散布図

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# i) csvデータ読込
df = pd.read_csv('../csv/003-1.csv')

# ii) pdだと扱いづらい → np配列に変換
# data = df.as_matrix() # Python3.7以降非推奨
data = df.values

# iii) データを分解
x = data[:,0]
y = data[:,1]

# iv) スキャタープロット描画
plt.scatter(x, y)
plt.show()

# v) フィットするラインを引こう
# ※本来は機械学習のアルゴリズムを用いてラインを模索する！（今回は割愛）
ans_x = np.linspace(0, 100 ,100)
ans_y = 2 * ans_x + 12
plt.scatter(x, y)
plt.plot(ans_x, ans_y, color='r', linewidth=3)
plt.show()
````

## ヒストグラム
---
> データの分布状況を把握するため

+ 標準正規分布 = Standard Normal Distribution
+ 標準正規分布は平均が0、分散が1の正規分布
+ 連続一様分布 = Continuous Uniform Distribution
+ np.random.randn() 標準正規分布に属する乱数生成
+ np.random.random() 連続一様分布に属する乱数生成

````python
import matplotlib.pyplot as plt
import numpy as np

# i) Standard Normal Distribution（標準正規分布:ガウス分布とも言う）
g = np.random.randn(1000)  # 1000個の乱数生成

# ii) ヒストグラム
plt.hist(g)
plt.show()

# iii) binを指定
# ※binを指定しなくとも自動で割り当ててくれる！
plt.hist(g, bins=100)  # 100個の階級指定
plt.show()

# iv) range指定
plt.hist(g, bins=100, range=(0, 5))
plt.show()

# v) 累積値
plt.hist(g, bins=100, cumulative=True)
plt.show()

# vi) 色や幅、塗りつぶしなんてのもできる！
plt.hist(g, bins=100, rwidth=0.5, color='red', histtype='step')
plt.show()

# vii) 2つのデータ同時プロッティング
# 連続一様分布を作成
n = np.random.random(1000)

labels = ['Gaussian', 'Continud Uniform Distribution']
plt.hist(g, bins=100, alpha=0.3, color='red')  #alpha：透明度
plt.hist(n, bins=100, alpha=0.3, color='blue')
plt.show()
````

## イメージプロッティング
---
> ※ MNISTのデータセットを読み込んで、画像のプロッティングを行っていきましょう。
>    KaggleからMNISTのデータセットをダウンロードして利用します。

### MNISTとは？
MNISTは28pix X 28pix＝784のピクセル情報。
「Mixed National Institute of Standards and Technology database」の略で、 手書きの数字「０〜９」の画像に対して、正解ラベルが付与されているデータセットを指します。
手軽に入手が可能で、またサイズ的にも扱いやすいことから、機械学習・人工知能の分野において初心者向けデータセットとして非常に有名です。

[MNISTデータダウロード先](https://www.kaggle.com/c/digit-recognizer/data)

https://www.kaggle.com/c/digit-recognizer/data?select=train.csv

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('../csv/train.csv')

# i) pd → np
A = df.values

# ii) サンプルでデータ取得 （784のピクセルデータをベクトル取得）
A1 = A[5, 1:]  # 5行目　０列目を除く全データ

# iii) 扱い易く ２８×２８ サイズに変更
A1 = A1.reshape(28, 28)

# iv) .imshow でプロット（image show）　
plt.imshow(A1)
plt.show()

# v) .imshow のオプション（グレースケール）
plt.imshow(A1, cmap='gray')
plt.show()

# vi) 別のデータを取得
A2 = A[10, 1:]
A2 = A2.reshape(28, 28)
plt.imshow(A2, cmap='gray')
plt.show()
````

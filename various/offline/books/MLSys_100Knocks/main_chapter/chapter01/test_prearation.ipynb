### ノック１：データを全て読み込んでみよう

```python
import pandas as pd
```

```python
# master data
m_area = pd.read_csv("input/m_area.csv")
m_store = pd.read_csv("input/m_store.csv")
# transaction data
tbl_order_4 = pd.read_csv("input/tbl_order_202004.csv")
tbl_order_5 = pd.read_csv("input/tbl_order_202005.csv")
tbl_order_6 = pd.read_csv("input/tbl_order_202006.csv")
```

### ノック２：データを結合（ユニオン）してみよう

```python
tbl_all = pd.concat([tbl_order_4, tbl_order_5, tbl_order_6], axis=0, ignore_index=True)
tbl_all.head()
```

```python
# 検算
len(tbl_all) == (len(tbl_order_4) + len(tbl_order_5) + len(tbl_order_6))
```

### ノック３：フォルダ内のファイル名を一覧化してみよう

```python
import os
import glob
```

```python
current_dir = os.getcwd() #カレントディレクトリ（現在位置）
os.listdir(current_dir)
```

```python
# tbl_orderに限って取得
path = os.path.join(current_dir, "input/tbl_order_*.csv") #取得したいパスを作成
tbl_order_files =  glob.glob(path)
tbl_order_files
```

### ノック４：複数データを結合（ユニオン）してみよう

```python
order_df = pd.DataFrame()

for i, file in enumerate(tbl_order_files):
    tbl_df = pd.read_csv(file)
    order_df = pd.concat([order_df, tbl_df], axis=0, ignore_index=True)
    print("{} {}: {}".format(i, file.split("/")[-1], len(order_df)))
```

```python
order_df.shape
```

### ノック６：不要なデータを除外しよう

```python
m_store.tail()
```

```python
len(order_df[order_df["store_id"]==999])
```

```python
order_df = order_df[order_df["store_id"] != 999]
order_df
```

### ノック７：マスタデータを結合（ジョイン）してみよう

```python
order_df = pd.merge(order_df, m_store, on="store_id", how="left")
order_df
```

```python
order_df = pd.merge(order_df, m_area, on="area_cd", how="left")
order_df
```

### ノック８：マスタが存在しないコードに名称を設定しよう

>|takeout_flag||
|:-:|:-|
|0|デリバリー|
|1|お持ち帰り|
>
>|status||
|:-:|:-|
|0|受付|
|1|お支払い済|
|2|お渡し済|
|9|キャンセル|

```python
order_df.loc[order_df["takeout_flag"] == 0, "takeout_name"] = "デリバリー"
```

```python
order_df.loc[order_df["takeout_flag"] == 1, "takeout_name"] = "お持ち帰り"
```

```python
order_df
```

```python
order_df["status"].unique()
```

```python
order_df.loc[order_df["status"] == 0, "status_name"] = "受付"
order_df.loc[order_df["status"] == 1, "status_name"] = "お支払済み"
order_df.loc[order_df["status"] == 2, "status_name"] = "お渡し済み"
order_df.loc[order_df["status"] == 9, "status_name"] = "キャンセル"
```

```python
order_df
```

### ノック９：分析基礎テーブルを出力してみよう

```python
output_dir = os.path.join(current_dir, "output")
os.makedirs(output_dir, exist_ok=True)
output_file = os.path.join(output_dir, "order_df.csv")
order_df.to_csv(output_file, index=False)
```
